```python
import torch
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt

torch.set_default_device("cuda")
```

# Output noise

```python
def solve_and_test(X,y,beta):
  beta_hat = torch.linalg.pinv(X) @ y

  return (beta_hat - beta).square().sum(), (y - X @ beta_hat).square().mean()

def theory(p,n):
  c = p/n
  if c > 1:
    return 1/(c-1)
  else:
    return c/(1-c)

def create_data(n,p,z):
  X = torch.zeros(n+p,p)
  X[:n,:] = torch.randn(n,p)/np.sqrt(p)
  for i in range(p):
    X[[n+i],:] = torch.rand(1)*z #+ torch.randn_like(z)*0.0001
  return X

def create_y(X,beta):
  y = X @ beta
  return y + torch.randn_like(y)

def compute_curve(p,z,beta,Ns,T):
  N = len(Ns)
  losses = torch.zeros(N)
  train_error = torch.zeros(N)
  for i in tqdm(range(N)):
    n = Ns[i]
    for j in range(T):
      X = create_data(n,p,z)
      y = create_y(X,beta)
      risk, train = solve_and_test(X,y,beta)
      losses[i] += risk/T
      train_error[i] += train/T
  return losses, train_error

def compute_theory_curve(p,min_N, max_N):
  errors = torch.zeros(max_N - min_N)
  Ns_theory = torch.zeros(max_N - min_N)
  N = max_N - min_N
  for i in range(N):
    errors[i] = theory(p,min_N+i+0.5)
    Ns_theory[i] = min_N+i+0.5 + p
  return errors, Ns_theory
```

```python
p = 300
z = torch.randn(1,p)/np.sqrt(p)
beta = torch.randn(p,1)
beta = beta #- z @ beta)

Ns = [250,275,325,350,400,450,500,550,600,650,700,750,800,850,900,950,1000,1050,1100]
T = 150
```

```python
loss, train = compute_curve(p,z,beta,Ns,T)
```

```python
errors, Ns_theory = compute_theory_curve(p,200,Ns[-1])
```

```python
plt.plot(p/(torch.tensor(Ns).cpu()+p),train.cpu(),linewidth= 3)
plt.xlabel(r"$\frac{d}{N_{trn}}$", fontsize = 18)
plt.ylabel(r"$\|\beta_{opt} - \beta\|^2$", fontsize = 18)
# plt.savefig("Output-noise-under.pdf", bbox_inches='tight', facecolor='white', dpi = 100, format = 'pdf')
```

```python
def solve_and_test(Xtrn, Xtst, n, p, mu, Q, beta):
  # A_hat = make_noise_ridge_rot(n,p,mu) #shift_spectrum(torch.randn(p,n)/np.sqrt(p),mu) #
  A_hat_2 = make_noise_ridge(n,p,mu)
  Atst = torch.randn_like(Xtst)/np.sqrt(p) #make_noise(Xtst.shape[1],p,mu, V=V)

  X_hat = torch.cat((Xtrn, torch.zeros(p,p)),dim = 1)
  # X_hat = Xtrn

  # W = X_hat @ torch.linalg.pinv(X_hat + A_hat)
  # W = X_hat @ torch.linalg.pinv(X_hat + A_hat)
  W_2 = beta @ X_hat @ Q @ torch.linalg.pinv(X_hat @ Q  + (A_hat_2))
  # return (Xtst - W @ (Xtst + Atst)).square().sum(), (Xtst - W_2 @ (Xtst + Atst)).square().sum()
  return W_2.square().sum().sqrt(), (beta @ Xtst - W_2 @ (Xtst + Atst)).square().sum()
```

```python
def compute_curve(u, mu, Ns, T, beta):
  p = u.shape[0]
  N = len(Ns)
  losses = torch.zeros(N)
  losses_2 = torch.zeros(N)
  vtst = torch.randn(1,Ns[-1])
  Xtst = u @ vtst
  for i,n in tqdm(list(enumerate(Ns))):
    vtrn = torch.randn(1,n)
    Xtrn = u @ vtrn

    for j in range(T):
      Q = torch.linalg.svd(torch.randn(n+p, n+p)/np.sqrt(n+p))[0] @ torch.linalg.svd(torch.randn(n+p, n+p)/np.sqrt(n+p))[2]
      # print(torch.norm(Q @ Q.T - torch.eye(n+p)))
      loss, loss_2 = solve_and_test(Xtrn, Xtst, n, p, mu, Q, beta)
      losses[i] += loss/T
      losses_2[i] += loss_2/T
  return losses, losses_2
```
